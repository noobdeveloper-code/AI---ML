**Title**: Understanding the Dataset <br>

**Task**: For the given datasets below, identify the data types and dimensions.<br>

Task 1: Employee Dataset<br>
Columns: Employee ID , Name , Age , Department , Salary , Joining Date<br>
Task 2: Product Sales Dataset<br>
Columns: Product ID , Product Name , Price , Quantity Sold , Category , Sales Date<br>
Task 3: Student Grades Dataset<br>
Columns: Student ID , Student Name , Math Score , Science Score , English Score ,Year<br>

Instructions:<br>

Identify which columns are numerical (continuous or discrete) and which are categorical
(nominal or ordinal).<br>
Note down the dimensions (number of rows and columns) of the dataset.

```python

import pandas as pd

# Sample data for Employee Dataset
employee_data = {
    "Employee ID": [101, 102, 103],
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Department": ["HR", "Finance", "IT"],
    "Salary": [50000, 60000, 70000],
    "Joining Date": ["2020-01-15", "2019-03-10", "2018-07-25"]
}
employee_df = pd.DataFrame(employee_data)

# Sample data for Product Sales Dataset
product_sales_data = {
    "Product ID": [1, 2, 3],
    "Product Name": ["Laptop", "Mouse", "Keyboard"],
    "Price": [1000, 50, 80],
    "Quantity Sold": [10, 200, 150],
    "Category": ["Electronics", "Accessories", "Accessories"],
    "Sales Date": ["2023-01-01", "2023-01-02", "2023-01-03"]
}
product_sales_df = pd.DataFrame(product_sales_data)

# Sample data for Student Grades Dataset
student_grades_data = {
    "Student ID": [1, 2, 3],
    "Student Name": ["John", "Jane", "Doe"],
    "Math Score": [85, 90, 95],
    "Science Score": [80, 85, 90],
    "English Score": [75, 80, 85],
    "Year": [2023, 2023, 2023]
}
student_grades_df = pd.DataFrame(student_grades_data)

# Function to identify data types and dimensions
def analyze_dataset(df, dataset_name):
    print(f"Dataset: {dataset_name}")
    print("Data Types:")
    print(df.dtypes)
    print(f"Dimensions: {df.shape[0]} rows, {df.shape[1]} columns")
    print("\n")

# Analyze each dataset
analyze_dataset(employee_df, "Employee Dataset")
analyze_dataset(product_sales_df, "Product Sales Dataset")
analyze_dataset(student_grades_df, "Student Grades Dataset")
```

**Title**: Checking for Missing Values<br>

**Task**: Identify and count the number of missing values in each dataset.<br>

Instructions:<br>
Use Python or any data manipulation tool to check for missing values in each column of the datasets. <br>Report the columns which have missing values and their counts.

```python
# Write your code from here


# Function to check for missing values in a dataset
def check_missing_values(df, dataset_name):
    print(f"Dataset: {dataset_name}")
    missing_values = df.isnull().sum()
    print(missing_values)
    print("Missing Values:")
    print(missing_values[missing_values > 0])  # Display only columns with missing values
    if missing_values.sum() == 0:
        print("No missing values found.")
    print("\n")

# Check for missing values in each dataset
check_missing_values(employee_df, "Employee Dataset")
check_missing_values(product_sales_df, "Product Sales Dataset")
check_missing_values(student_grades_df, "Student Grades Dataset")
```

**Title**: Handling Outliers<br>

**Task**: Detect and propose handling methods for outliers in the numerical columns of the datasets.<br>

Task 1: Age in Employee Dataset<br>
Task 2: Price in Product Sales Dataset<br>
Task 3: Math Score in Student Grades Dataset<br>

Instructions:<br>

Use box plots to visualize potential outliers.<br>
Suggest methods to handle them, such as removal or transformation.

```python
# Write your code from here
import matplotlib.pyplot as plt
import seaborn as sns

# Function to create box plots for outlier detection
def visualize_outliers(df, column, dataset_name):
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[column], color='skyblue')
    plt.title(f'Box Plot of {column} in {dataset_name}')
    plt.ylabel(column)
    plt.show()

# Task 1: Age in Employee Dataset
visualize_outliers(employee_df, "Age", "Employee Dataset")

# Task 2: Price in Product Sales Dataset
visualize_outliers(product_sales_df, "Price", "Product Sales Dataset")

# Task 3: Math Score in Student Grades Dataset
visualize_outliers(student_grades_df, "Math Score", "Student Grades Dataset")
```

**Title**: Finding Relationships Between Features<br>

**Task**: Identify relationships between pairs of features in the datasets.<br>

Task 1: Salary vs Age in Employee Dataset<br>
Task 2: Price vs Quantity Sold in Product Sales Dataset<br>
Task 3: Math Score vs Science Score in Student Grades Dataset

Instructions: <br>

Use scatter plots or correlation coefficients to analyze the relationships.<br>
Describe any insights or patterns observed.

```python
# Write your code from here
import matplotlib.pyplot as plt
import seaborn as sns

# Function to create scatter plots and calculate correlation coefficients
def analyze_relationship(df, x_column, y_column, dataset_name):
    # Scatter plot
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[x_column], y=df[y_column], color='blue', s=100)
    plt.title(f'{y_column} vs {x_column} in {dataset_name}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
    
    # Correlation coefficient
    correlation = df[[x_column, y_column]].corr().iloc[0, 1]
    print(f"Correlation between {x_column} and {y_column} in {dataset_name}: {correlation:.2f}\n")

# Task 1: Salary vs Age in Employee Dataset
analyze_relationship(employee_df, "Age", "Salary", "Employee Dataset")

# Task 2: Price vs Quantity Sold in Product Sales Dataset
analyze_relationship(product_sales_df, "Price", "Quantity Sold", "Product Sales Dataset")

# Task 3: Math Score vs Science Score in Student Grades Dataset
analyze_relationship(student_grades_df, "Math Score", "Science Score", "Student Grades Dataset")
```

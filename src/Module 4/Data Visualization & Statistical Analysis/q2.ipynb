**Multivariate Analysis (Higher-Dimensional Data Relationships)**<br>
Objective: Analyze relationships in higher-dimensional data using pair plots and PCA.

Title: Pair Plots<br>

Task 1: Create a pair plot for the Iris dataset.<br>
Task 2: Focus on a subset of features (e.g., only petal dimensions).<br>
Task 3: Exclude one class to observe differences.

```python
# Write your code from here
# Import necessary libraries
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Task 1: Create a pair plot for the entire Iris dataset
sns.pairplot(df, hue='species', diag_kind='kde', corner=True, palette='Set2')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# Task 2: Focus on a subset of features (e.g., only petal dimensions)
sns.pairplot(df[['petal length (cm)', 'petal width (cm)', 'species']], hue='species', diag_kind='kde', corner=True, palette='Set2')
plt.suptitle('Pair Plot of Petal Dimensions (Iris Dataset)', y=1.02)
plt.show()

# Task 3: Exclude one class to observe differences
# Filter out one class (e.g., species 0)
df_filtered = df[df['species'] != 0]
sns.pairplot(df_filtered, hue='species', diag_kind='kde', corner=True, palette='Set2')
plt.suptitle('Pair Plot Excluding Species 0 (Iris Dataset)', y=1.02)
plt.show()
```

Title: Principal Component Analysis (PCA)<br>

Task 1: Perform PCA and plot the first two principal components.<br>
Task 2: Visualize explained variance by each component.<br>
Task 3: Retain more components and visualize in 3D.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (e.g., x1, x2, x3, x4)
y = iris.target  # Target labels

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Task 1: Perform PCA and plot the first two principal components
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

# Create a DataFrame for visualization
pca_df_2d = pd.DataFrame(data=X_pca_2d, columns=['Principal Component 1', 'Principal Component 2'])
pca_df_2d['Target'] = y

# Plot the first two principal components
plt.figure(figsize=(8, 6))
for target, color in zip([0, 1, 2], ['red', 'green', 'blue']):
    subset = pca_df_2d[pca_df_2d['Target'] == target]
    plt.scatter(subset['Principal Component 1'], subset['Principal Component 2'], label=iris.target_names[target], color=color, alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Principal Components')
plt.legend()
plt.grid()
plt.show()

# Task 2: Visualize explained variance by each component
pca_full = PCA()
pca_full.fit(X_scaled)
explained_variance = pca_full.explained_variance_ratio_

plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.show()

# Task 3: Retain more components and visualize in 3D
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)

# Create a DataFrame for 3D visualization
pca_df_3d = pd.DataFrame(data=X_pca_3d, columns=['PC1', 'PC2', 'PC3'])
pca_df_3d['Target'] = y

# Plot the first three principal components in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
for target, color in zip([0, 1, 2], ['red', 'green', 'blue']):
    subset = pca_df_3d[pca_df_3d['Target'] == target]
    ax.scatter(subset['PC1'], subset['PC2'], subset['PC3'], label=iris.target_names[target], color=color, alpha=0.7)
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA: First Three Principal Components')
ax.legend()
plt.show()
```

**Statistical Analysis in EDA**<br>

Objective: Calculate basic statistical metrics and explore the relationship between features using correlation and covariance.<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from scipy.stats import skew, kurtosis

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Task 1: Calculate mean, median, and standard deviation of petal length
petal_length = df['petal length (cm)']
petal_length_mean = petal_length.mean()
petal_length_median = petal_length.median()
petal_length_std = petal_length.std()

print(f"Petal Length - Mean: {petal_length_mean:.2f}, Median: {petal_length_median:.2f}, Standard Deviation: {petal_length_std:.2f}")

# Task 2: Calculate skewness and kurtosis of sepal width
sepal_width = df['sepal width (cm)']
sepal_width_skewness = skew(sepal_width)
sepal_width_kurtosis = kurtosis(sepal_width)

print(f"Sepal Width - Skewness: {sepal_width_skewness:.2f}, Kurtosis: {sepal_width_kurtosis:.2f}")

# Task 3: Calculate mean, median, and standard deviation of sepal length
sepal_length = df['sepal length (cm)']
sepal_length_mean = sepal_length.mean()
sepal_length_median = sepal_length.median()
sepal_length_std = sepal_length.std()

print(f"Sepal Length - Mean: {sepal_length_mean:.2f}, Median: {sepal_length_median:.2f}, Standard Deviation: {sepal_length_std:.2f}")
```

Title: Correlation & Covariance<br>

Task 1: Compute correlation between sepal length and petal length.<br>
Task 2: Compute covariance between petal width and sepal width.<br>
Task 3: Determine the most correlated pair of features.<br>

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Task 1: Compute correlation between sepal length and petal length
sepal_petal_corr = df['sepal length (cm)'].corr(df['petal length (cm)'])
print(f"Correlation between Sepal Length and Petal Length: {sepal_petal_corr:.2f}")

# Task 2: Compute covariance between petal width and sepal width
petal_sepal_cov = df['petal width (cm)'].cov(df['sepal width (cm)'])
print(f"Covariance between Petal Width and Sepal Width: {petal_sepal_cov:.2f}")

# Task 3: Determine the most correlated pair of features
correlation_matrix = df.corr()
most_correlated_pair = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates()
most_correlated_pair = most_correlated_pair[most_correlated_pair < 1].idxmax()
most_correlation_value = correlation_matrix.loc[most_correlated_pair]
print(f"Most correlated pair of features: {most_correlated_pair} with correlation {most_correlation_value:.2f}")
```

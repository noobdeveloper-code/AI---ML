Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Drop missing values in the 'age' column
titanic = titanic.dropna(subset=['age'])

# Create a histogram for the 'age' feature
plt.hist(titanic['age'], bins=15, color='blue', alpha=0.7, edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age (Titanic Dataset)')

# Display the histogram
plt.show()
```

Title: Correlation Heatmap<br>

Question 5 : Generate a correlation heatmap for a dataset with multiple features (e.g., height ,weight , age ) and explain the correlations observed.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load the 'bodyfat' dataset from OpenML
bodyfat = fetch_openml(name='bodyfat', version=1, as_frame=True)
df = bodyfat.frame

# Select relevant features (e.g., height, weight, age)
df = df[['Height', 'Weight', 'Age']].dropna()

# Compute the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add a title
plt.title('Correlation Heatmap (Bodyfat Dataset)')

# Display the heatmap
plt.show()
```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
from scipy.stats import kurtosis, skew

# Simulate a dataset with a 'temperature' feature for demonstration purposes
data = {'temperature': [72, 75, 78, 80, 85, 90, 95, 100, 105, 110]}
df = pd.DataFrame(data)

# Calculate descriptive statistics
mean_temp = df['temperature'].mean()
median_temp = df['temperature'].median()
std_temp = df['temperature'].std()
skew_temp = skew(df['temperature'])
kurt_temp = kurtosis(df['temperature'])

# Print the results
print(f"Mean: {mean_temp}")
print(f"Median: {median_temp}")
print(f"Standard Deviation: {std_temp}")
print(f"Skewness: {skew_temp}")
print(f"Kurtosis: {kurt_temp}")
```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd

# Simulate a dataset with 'price' and 'demand' features
data = {
    'price': [10, 15, 20, 25, 30],
    'demand': [100, 80, 60, 40, 20]
}
df = pd.DataFrame(data)

# Compute covariance
covariance = df['price'].cov(df['demand'])

# Compute correlation
correlation = df['price'].corr(df['demand'])

# Print the results
print(f"Covariance between price and demand: {covariance}")
print(f"Correlation between price and demand: {correlation}")
```

Title: Pair Plot for Multivariate Analysis<br>

Question 8: Utilize a pair plot on a dataset to explore the relationships and distributions between
height , weight , and age . What insights can you glean from this visualization?

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load the 'bodyfat' dataset from OpenML
bodyfat = fetch_openml(name='bodyfat', version=1, as_frame=True)
df = bodyfat.frame

# Select relevant features (e.g., height, weight, age)
df = df[['Height', 'Weight', 'Age']].dropna()

# Create a pair plot for the selected features
sns.pairplot(df, diag_kind='kde', corner=True, plot_kws={'alpha': 0.7})

# Add a title
plt.suptitle('Pair Plot of Height, Weight, and Age (Bodyfat Dataset)', y=1.02)

# Display the pair plot
plt.show()
```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (e.g., x1, x2, x3, x4)
y = iris.target  # Target labels

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to two principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for visualization
pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Target'] = y

# Visualize the data in the new feature space
plt.figure(figsize=(8, 6))
for target, color in zip([0, 1, 2], ['red', 'green', 'blue']):
    subset = pca_df[pca_df['Target'] == target]
    plt.scatter(subset['Principal Component 1'], subset['Principal Component 2'], label=iris.target_names[target], color=color, alpha=0.7)

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Iris Dataset (2 Principal Components)')
plt.legend()
plt.grid()

# Display the plot
plt.show()
```

Title: Advanced Pair Plot with Hue Parameter<br>

Question 10 : Create a pair plot for height , weight , and age with an added categorical variable gender as the hue to observe different group trends.

```python
# Write your code from here
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Load the 'bodyfat' dataset from OpenML
bodyfat = fetch_openml(name='bodyfat', version=1, as_frame=True)
df = bodyfat.frame

# Simulate a 'Gender' column for demonstration purposes
import numpy as np
np.random.seed(42)
df['Gender'] = np.random.choice(['Male', 'Female'], size=len(df))

# Select relevant features (e.g., height, weight, age, gender)
df = df[['Height', 'Weight', 'Age', 'Gender']].dropna()

# Create a pair plot with 'Gender' as the hue
sns.pairplot(df, hue='Gender', diag_kind='kde', corner=True, plot_kws={'alpha': 0.7})

# Add a title
plt.suptitle('Pair Plot of Height, Weight, and Age with Gender (Bodyfat Dataset)', y=1.02)

# Display the pair plot
plt.show()
```
